#### Converting required columns to integer

```python
df["km"] = df['km'].str.replace(',', '').astype(int)
df["emi"] = df["emi"].str.replace(',','').astype(int)
df["price"] = df["price"].str.replace(',','').astype(int)
```

#### Piechart

```python
#fuel
plt.figure(figsize = (7,7))
wp = {'linewidth':2, 'edgecolor': "black"}
explode = (0.1,0.1,0.1)
tags = df["fuel"].value_counts()
tags.plot(kind = "pie",autopct = '%1.1f%%',shadow = True,startangle = 90, wedgeprops = wp,
          explode = explode, label = '')
plt.title("Distribution of different Fuel types",fontsize=18)
plt.show()
```

#### Histogram of numerical data

#### Distance plot for numerical data

```python
num = df[["year","km","emi","price"]]
for i in num:
    plt.figure(figsize = (10,7))
    sns.distplot(num[i])
    plt.title(i)
    plt.show()
```

#### Boxplot

```python
for i in num:
    plt.figure(figsize = (12,2))
    sns.boxplot(num[i])
    plt.title(i)
    plt.show()
```

#### Scatterplot for numerical data

```python
for i in num:
    plt.figure(figsize= (10,7))
    sns.scatterplot(x = num[i],y = df["price"],hue = df["transmission"])
    plt.xlabel(i)
    plt.ylabel("Price")
    plt.title(f"{i} vs price")
    plt.show()        
```

#### Replacing  some categorical values into integer
